```python
run_8 = run[run['Unnamed: 0']%8==0]
run_8['Unnamed: 0'] = run_8['Unnamed: 0']/8
run_8.head()
```

```python
total = run_8.filter(regex='total_|Unnamed').drop(columns='total_genco_wealth')
total.head()
total_long = total.melt(id_vars=['Unnamed: 0'])
sns.lineplot(data=total_long, x='Unnamed: 0', y='value', hue='variable')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
