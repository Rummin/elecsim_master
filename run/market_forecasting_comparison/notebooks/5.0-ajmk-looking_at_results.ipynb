```python
def get_metric(results_dict, metric):
    hold_dat = {}
    for res in results_dict:
        hold_frame = []
        for i in range(len(results_dict['{"C": 0.1, "fit_intercept": true, "max_iter": 1, "shuffle": false, "tol": 0.001}'])):
             hold_frame.append(results_dict[res][i][metric])
        hold_dat[res] = hold_frame

    results_df = pd.DataFrame(hold_dat)
    return results_df
```

```python
# def get_metric_season(results_dict, metric):
hold_dat = {}
for res in results_dict:
    hold_frame = []
    for i in range(24):
#         print(results_dict[res].values[0][i]['mean_absolute_error'])
        hold_frame.append(results_dict[res].values[0][i]['mean_absolute_error'])
        #          hold_frame.append(results_dict[res][i][metric])
    hold_dat[res] = hold_frame

results_df = pd.DataFrame(hold_dat)
results_df
#     return results_df
```

```python
results_df.mean()
```

```python
results_df_transposed = results_df.transpose().reset_index()

results_df_no_duplicates = results_df_transposed.drop_duplicates(subset=[0,1,2,3,4,5,6,7])

results_df_transposed_long = results_df_no_duplicates.melt(id_vars='index', value_vars=list(range(0,24)))
results_df_transposed_long
```

```python
results_df_transposed_long['index'] = results_df_transposed_long['index'].str.replace(":","=").str.replace('{',"").str.replace('"',"").str.replace("="," =").str.replace(", solver = adam}", "").str.replace(", max_iter = 1, shuffle = false, tol = 0.001","").str.replace("}","")
results_df_transposed_long
results_df_transposed_long['index'] = results_df_transposed_long['index'].astype(str)
results_df_transposed_long



results_df_transposed_long['index'] = np.where(results_df_transposed_long['index'].str.contains("hidden_layer_"),'(MLP): ' + results_df_transposed_long['index'], results_df_transposed_long['index'])
results_df_transposed_long['index'] = np.where(results_df_transposed_long['index'].str.contains("C ="),'(PA): ' + results_df_transposed_long['index'], results_df_transposed_long['index'])
results_df_transposed_long['index'] = np.where(results_df_transposed_long['index'].str.contains("power"),'(Box Cox): ' + results_df_transposed_long['index'], results_df_transposed_long['index'])
results_df_transposed_long['index'].iloc[6] = "Linear Regression"
results_df_transposed_long['index'] = np.where(results_df_transposed_long['index'].str.contains("(MLP)"),'MLP (all parameter variations)', results_df_transposed_long['index'])


results_df_transposed_long.replace("", float("NaN"), inplace=True)

results_df_transposed_long.dropna(subset = ["index"], inplace=True)
```
